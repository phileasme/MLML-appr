```python
# We set the distance to k20 and store the results.
# from KNearestNeighbors import KNearestNeighbor, classification_accuracy
import pandas as pd
import numpy as np
from tqdm import tqdm
import torch
import pickle
```

```python
x_s, y_s = [pd.read_pickle(pkl) for pkl in ['./x_s.pkl', './y_s.pkl']]
```

```python
slicing = int(1*(len(y_s)))
x_s_train, y_s_train = x_s.iloc[:slicing,:], y_s.iloc[:slicing,:]
# x_s_test, y_s_test  = x_s.iloc[slicing:,:], y_s.iloc[slicing:,:]
```

```python
xs_train_np = x_s_train.to_numpy()
ys_train_np = y_s_train.to_numpy()
# xs_test_np = x_s_test.to_numpy()
# ys_test_np = y_s_test.to_numpy()
```

```python
xs_train_np.shape
```

```python
b = np.array([range(xs_train_np.shape[0])])
xs_train_with_idx = np.concatenate((xs_train_np,  b.T), axis=1)
int(xs_train_with_idx[0][-1])
```

```python
# Currently this wouldn't work as it is set for single label problem.
# knn = KNearestNeighbor(xs_train_np, ys_train_np)
# Okay first thing, add their actually indices as a column
# Then compute the ranks for each one of them.
# Store each up to 20th rank 
```

Computing $ V_X$, $V_{X}(i, j)=\exp \left(-d^{2}\left(x_{i}, x_{j}\right) / \sigma_{i} \sigma_{j}\right)$

```python
rank_one_array = []
xs_train_20_nearest = []
xs_train_7th = []
V_X = []
# Ignore the weights, simply use Euclidian distance

for current in xs_train_with_idx:
    order_values = [(None, None)] * 20
    for other in xs_train_with_idx:
        if current[-1] != other[-1]:
            euclidean = (current[:-1] - other[:-1])**2
            d_square = sum(euclidean)
            idx = other[-1]
            shifting = False
            temporary = (idx, d_square)
            for i in range(len(order_values)):
                if shifting:
                    current_temp = order_values[i]
                    order_values[i] = temporary
                    temporary = current_temp
                    continue
                if order_values[i][0] == None and not shifting:
                    order_values[i] = (idx, d_square)
                    break
                if order_values[i][1] >= d_square:
                    temporary = order_values[i]
                    order_values[i] = (idx, d_square)
                    shifting = True
    xs_train_7th.append(order_values[6])
    xij = [0.0] * xs_train_with_idx.shape[0]
    for e in order_values:
        xij[int(e[0])] = e[1]
    V_X.append(xij)
    xs_train_20_nearest.append([round(e[1],4) for e in order_values])
V_X = np.array(V_X)
for i in tqdm(range(V_X.shape[0])):
    for j in range(V_X.shape[1]):
        V_X[i][j] /= (xs_train_7th[i][1] * xs_train_7th[j][1] )
V_X = np.exp(-V_X)
```

Computing $D_{X}=\operatorname{diag}\left(d_{X}(1), \cdots, d_{X}(n)\right)$ where $d_{X}(i)=\sum_{j}^{n} V_{X}(i, j)$

```python
V_X = torch.from_numpy(V_X).float()
D_X = []
for v_x_i in V_X:
    D_X.append(sum(v_x_i))
D_X = torch.diag(torch.Tensor(D_X).float())
```

$ L_{X}=I-D_{X}^{-\frac{1}{2}} V_{X} D_{X}^{-\frac{1}{2}} $

```python
D_X_sqrt_inverse = torch.sqrt(torch.inverse(D_X))
L_X = torch.eye(D_X.shape[0]) - torch.mm(D_X_sqrt_inverse, torch.mm(V_X, D_X_sqrt_inverse))
L_X.shape
L_X_fileObject = open('./L_X.pkl','wb')
pickle.dump(L_X, L_X_fileObject)  
```
